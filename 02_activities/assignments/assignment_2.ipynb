Create two [Column Transformers](https://scikit-learn.org/stable/modules/generated/sklearn.compose.ColumnTransformer.html), called preproc1 and preproc2, with the following guidelines:

+ The only difference between preproc1 and preproc2 is the non-linear transformation of the numerical variables.
    


### Preproc 1

Create preproc1 below.

+ Numeric: scaled variables, no other transforms.
+ Categorical: one-hot encoding.

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder

numeric = [
    'coord_x', 'coord_y', 'ffmc', 'dmc',
    'dc', 'isi', 'temp', 'rh', 'wind', 'rain'
]
categorical = ['month', 'day']

preproc1 = ColumnTransformer(transformers=[
    ('num', StandardScaler(), numeric),
    ('cat', OneHotEncoder(handle_unknown='ignore'), categorical)
])
```

### Preproc 2

Create preproc1 below.

+ Numeric: scaled variables, non-linear transformation to one or more variables.
+ Categorical: one-hot encoding.

```python
temp = ['temp']
other_numeric = [f for f in numeric if f not in temp]

preproc2 = ColumnTransformer([
    ('num', StandardScaler(), other_numeric),
    ('power', Pipeline([
        ('scale', StandardScaler()),
        ('power', PowerTransformer())
    ]), temp),
    ('cat', OneHotEncoder(handle_unknown='ignore'), categorical)
])
```

## Regressor

+ Use a regression model to perform a prediction. 

    - Choose a baseline regressor, tune it (if necessary) using grid search, and evaluate it using cross-validation.
    - Choose a more advance regressor, tune it (if necessary) using grid search, and evaluate it using cross-validation.

```python
# Pipeline A = preproc1 + baseline

from sklearn.linear_model import Ridge  # or your choice of regressor

pipeA = Pipeline([
    ('preprocessing', preproc1),
    ('regressor', Ridge())
])
```

```python
# Pipeline B = preproc2 + baseline

pipeB = Pipeline([
    ('preprocessing', preproc2),
    ('regressor', Ridge())
])
```

```python
# Pipeline C = preproc1 + advanced model
pipeC = Pipeline([
    ('preprocessing', preproc1),
    ('regressor', RandomForestRegressor(random_state=42))
])
```

```python
# Pipeline D = preproc2 + advanced model

pipeD = Pipeline([
    ('preprocessing', preproc2),
    ('regressor', RandomForestRegressor(random_state=42))
])    
```

# Tune Hyperparams

+ Perform GridSearch on each of the four pipelines. 
+ Tune at least one hyperparameter per pipeline.
+ Experiment with at least four value combinations per pipeline.

```python
import numpy as np
from sklearn.pipeline import Pipeline
from sklearn.linear_model import Ridge
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV



# Hyperparameters
param_grid_ridge = {
    'regressor__alpha': np.logspace(-4, 2, 13),    
    'regressor__fit_intercept': [True, False]
}

param_grid_rf = {
    'regressor__n_estimators': [50, 100, 200, 500, 1000],
    'regressor__max_depth': [None, 5, 10, 20, 30]
}

# GridSearchCV
searchA = GridSearchCV(pipeA, param_grid_ridge, cv=5,
                       scoring='neg_mean_squared_error',
                       error_score='raise',
                       n_jobs=-1)
searchB = GridSearchCV(pipeB, param_grid_ridge, cv=5,
                       scoring='neg_mean_squared_error',
                       error_score='raise',
                       n_jobs=-1)
searchC = GridSearchCV(pipeC, param_grid_rf,    cv=5,
                       scoring='neg_mean_squared_error',
                       error_score='raise',
                       n_jobs=-1)
searchD = GridSearchCV(pipeD, param_grid_rf,    cv=5,
                       scoring='neg_mean_squared_error',
                       error_score='raise',
                       n_jobs=-1)

# Fitting all four
for name, search in zip(['A','B','C','D'], [searchA, searchB, searchC, searchD]):
    print(f"\nFitting pipeline {name}...")
    search.fit(X, y)
    best_mse  = -search.best_score_       # MSE
    best_rmse = np.sqrt(best_mse)         # RMSE
    print(f"  Best params: {search.best_params_}")
    print(f"  CV RMSE     : {best_rmse:.3f}")

```
